### __Python DataVisualization__

Why visualize data?
There are two fundamental reasons to create visualizations as a data professional:

- To simplify the analysis of large amounts of data by visually displaying its properties. In other words, visualize to analyze. By visualizing data, you can gain insights that would be impossible if you were reading raw numbers.

- To communicate the results of your analysis to your colleagues or clients in an understandable and effective way.
Data represented visually is often much easier to understand than raw data. Visualizations can also generate an emotional response, which can be important and very useful when trying to communicate your work.

##### _Elements of a Good Chart_

Good charts help us easily draw conclusions from the data. They give us a good initial understanding of the data without the need for in-depth analysis. However, the following chart isn't very useful for this purpose, as it's not clear what the data represents. Let's review some elements we could add to this chart to improve it.

##### _Titles_

Each graph should have a descriptive title that immediately communicates to the audience why it was created.

Titles shouldn't simply reiterate what the axis labels already communicate. Look at the mouse body weight graph, for example. It's titled "Growth Rate of Mice (Diet #1)," which is much better than a title like "Mouse Body Weight vs. Age." It's clear from the axis labels that we're plotting body weight vs. age, so a title like that doesn't provide any additional information.

Titles can even include a subtitle explaining the graph's conclusion. For example, "Growth Rate of Mice (Diet #1): The weight of mice grows linearly with age until about 4 months, after which their weight begins to stabilize."

##### _Axes_

Axes should always be labeled to avoid any ambiguity about the quantities or categories being graphed.

If an axis represents a physical quantity (e.g., weight, speed, time, money), you should always include the units for that quantity in the axis label. You should also include axis markers with numerical or categorical labels to give your audience an idea of ​​the scale of the data.

Typically, the best option is to plot the independent variable on the X-axis, while the dependent variable is plotted on the Y-axis.

The independent variable is the cause of the variation in the graph. Its value is independent of other variables in your graph.
The dependent variable is the effect. Its value depends on changes in the independent variable.

__(X - Ind / cause, Y - Dep / effect)__

##### _Legends_
Legends are useful when you plot more than one set of data on the same graph.

The mouse graph includes a color-coded legend that tells the audience that there is data for two different mice, essentially allowing us to plot a third variable (mouse number) on a 2D graph by using colors to differentiate the mice. The legend communicates the meaning of each color (purple for mouse 1 and green for mouse 2).

Not all charts require a legend, but a legend can help reiterate information for some readers.

##### __Creating Plots in Python with Matplotlib__

Python has several popular libraries for creating plots from your data. Perhaps the most common library is Matplotlib. So you'll likely find many references to this library on the web when searching for how to solve plotting problems in Python.

More recently, other more user-friendly plotting libraries have been created using Matplotlib as a base. In particular, Seaborn and Plotly are common choices. Plotly even has a newer library called Plotly Express, which is very user-friendly and often produces attractive plots right out of the box.

However, for many types of simple plots, we don't need an external plotting library at all! The good, trusty Pandas has its own methods for creating plots from Series and DataFrames objects. In this lesson, you'll learn how to create plots using Pandas and how to display them using Matplotlib.

##### The _plot()_ and _plt.show()_ Methods

Let's start by learning about the pandas plot() method and the show() method in Matplotlib's pyplot library. Here's an example of how we can call plot() to create a plot in pandas and then use show() to clearly display the plot.

```python
import pandas as pd
from matplotlib import pyplot as plt

df = pd.DataFrame({'a':[2, 3, 4, 5], 'b':[4, 9, 16, 25]})
print(df)

df.plot()
plt.show()
```

To draw this graph, we imported both pandas and pyplot from the matplotlib libraries. A popular convention is to use plt as the alias for the pyplot library.

The pandas plot() method prepares the image with the data, while the pyplot show() function presents the image to the user.

By default, the plot() method creates a graph where all numeric columns in the DataFrame are plotted on the same axes. The indices are on the X axis, and the column values ​​are on the Y axis.

If necessary, you can also plot only an individual column.

```python
import pandas as pd
from matplotlib import pyplot as plt

df = pd.DataFrame({'a':[2, 3, 4, 5], 'b':[4, 9, 16, 25]})

df['b'].plot()
plt.show()
```

In some development environments, such as Jupyter notebooks, you don't need to use plt.show() to display your plot. However, it's good practice to always include plt.show() so that your plots display in any environment where someone else might run your code.

Another option you might need is to write code that stores our plots as image files. For example, we can save the above plot in PNG format using matplotlib's savefig() function:

```python
import pandas as pd
from matplotlib import pyplot as plt

df = pd.DataFrame({'a':[2, 3, 4, 5], 'b':[4, 9, 16, 25]})

df['b'].plot()
#plt.savefig('myplot.png')
```

If you run this code on your local machine, you should have a PNG file named myplot.png in whatever directory your Python script or Jupyter notebook is in. Python knows what format to save the file in based on the extension in your string input to savefig().

##### _Customizing Charts with Plot() Parameters_

You may have noticed that the charts we've created so far are quite basic and don't adhere to the best practices you learned earlier in this chapter. Well, plot() has many additional parameters we can use to fix this.

_plot()_: Pandas method that prepares the image with the data.

_show()_: Matplotlib pyplot library function that presents the image to the user.

_savefig()_: Matplotlib function that stores the graph as an image file.

_title=_: Parameter to set the graph title.

_style=_: Parameter to change the graph marker style.

_x= and y=_: Parameters to plot two columns, one opposite the other.

_xlabel= and ylabel=_: Parameters to set the axis labels.

_xlim= and ylim=_: Parameters to set the axis limits.

_grid=_: Parameter to add grid lines.

_figsize=_: Parameter to add the graph size.

_legend=_: Parameter to add or remove the legend.

.

_Titles_

Chart titles are passed to the title= parameter as strings:

```python
import pandas as pd
from matplotlib import pyplot as plt

df = pd.DataFrame({'a':[2, 3, 4, 5], 'b':[4, 9, 16, 25]})

df.plot(title='A y B')
plt.show()
```

##### _Line Styles_

Our chart isn't that bad, but a line chart might not be the best option here, as it implies we have data points along the entire line when, in fact, we only have four. To make it obvious to anyone looking at the chart, it would be better to use a different marking style for the data points.

To change the chart's marker style, you can use the _style=_ parameter. We'll pass it an argument of _'o'_ so that each value is marked as a point

```python
import pandas as pd
from matplotlib import pyplot as plt

df = pd.DataFrame({'a':[2, 3, 4, 5], 'b':[4, 9, 16, 25]})

df.plot(title='A y B', style='o')
plt.show()
```

There are other marker styles as well; for example, style='х' marks each point with an x:

```python
import pandas as pd
from matplotlib import pyplot as plt

df = pd.DataFrame({'a':[2, 3, 4, 5], 'b':[4, 9, 16, 25]})

df.plot(title='A y B', style='x')
plt.show()
```

You can even combine different styles. For example, if you want to display both lines and dots, you can use style='o-':

```python
import pandas as pd
from matplotlib import pyplot as plt

df = pd.DataFrame({'a':[2, 3, 4, 5], 'b':[4, 9, 16, 25]})

df.plot(title='A y B', style='o-')
plt.show()
```

##### _Axes_

In all the charts you've seen so far, the horizontal axis contains the values ​​from the DataFrame index. But what if you want to plot two columns side by side? We can use the x= and y= parameters to accomplish this:

```python
import pandas as pd
from matplotlib import pyplot as plt

df = pd.DataFrame({'a':[2, 3, 4, 5], 'b':[4, 9, 16, 25]})

df.plot(x='b', y='a', title='A vs B', style='o')
plt.show()
```

##### _Legend_

Here, column 'a' is on the Y-axis and column 'b' is on the X-axis. Notice that pandas has automatically renamed the horizontal axis 'b', and the default legend now refers only to column 'a'.

To change the axis labels, we can use the xlabel= and ylabel= parameters of the plot() method:

```python
import pandas as pd
from matplotlib import pyplot as plt

df = pd.DataFrame({'a':[2, 3, 4, 5], 'b':[4, 9, 16, 25]})

df.plot(x='b',
        y='a',
        title='A vs B',
        style='o',
        xlabel="Hola, soy B",
        ylabel="Hola, soy A")

plt.show()
```

It's important to mention that there is an alternative way to set the labels for both the X and Y axes. Instead of specifying xlabel= and ylabel= in the plot() method, we can set the labels via plt.xlabel() and plt.ylabel(). Here's how to do it:

```python
import pandas as pd
from matplotlib import pyplot as plt

df = pd.DataFrame({'a':[2, 3, 4, 5], 'b':[4, 9, 16, 25]})

df.plot(x='b',
        y='a',
        title='A vs B',
        style='o',
        legend=False)

plt.xlabel("Hola, soy B") # Configuring legend x
plt.ylabel("Hola, so A") # Configuring legend y
plt.show()
```

Next, we can also set the axis limits using the xlim= and ylim= parameters. These take a number or a list of two numbers as their arguments. If you pass a single number, it will be the minimum value displayed for the corresponding axis. If you pass a list of two numbers, the first number will be the minimum value and the second number will be the maximum.

Let's expand the range of the horizontal axis from 0 to 30, and set the minimum value of the vertical axis to 0:

```python
import pandas as pd
from matplotlib import pyplot as plt

df = pd.DataFrame({'a':[2, 3, 4, 5], 'b':[4, 9, 16, 25]})

df.plot(x='b', y='a', title='A vs B', style='o', xlim=[0, 30], ylim=0)
plt.show()
```

##### _Other customization options_

You can add other features to your charts if you think they'll make them easier to read. Let's add grid lines to help the eye see what values ​​we're plotting. To do this, set the grid= parameter to True (by default, this parameter is set to False):

```python
import pandas as pd
from matplotlib import pyplot as plt

df = pd.DataFrame({'a':[2, 3, 4, 5], 'b':[4, 9, 16, 25]})

df.plot(x='b', y='a', title='A vs B', style='o', grid=True)
plt.show()
```

We can also manage the size of our chart with the figsize= parameter. The width and height in inches are passed as a list: figsize=[width, height]. Let's compare two charts of different sizes:

```python
import pandas as pd
from matplotlib import pyplot as plt

df = pd.DataFrame({'a':[2, 3, 4, 5], 'b':[4, 9, 16, 25]})

# generate a small graph
df.plot(x='b', y='a', style='o', xlim=[0, 30], figsize=[2, 2])

# generate a big graph
df.plot(x='b', y='a', style='o', xlim=[0, 30], figsize=[10, 4])

plt.show()
```
